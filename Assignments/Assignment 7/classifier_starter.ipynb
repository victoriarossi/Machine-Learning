```python
import numpy as np
import matplotlib.pyplot as plt 
import pickle

pkl_file = open('classifier_data.pkl', 'rb')
x_train, y_train = pickle.load(pkl_file)

n_train = np.size(y_train)

plt.scatter(x_train[:,0],x_train[:,1], c=y_train[:,0])
plt.title('training data')
plt.show()
```

```python
def regularized_log_loss(_w):
    assert np.shape(X)[1]==len(y)
    assert np.shape(X)[0]==np.shape(_w)[0]
    _nnl = 0
    for i in range(len(y)):
        _nnl += np.log(1 + np.exp(-y[i]*(X[:,[i]].T@_w)))[0,0]
    return _nnl + np.linalg.norm(_w)

def grad(_w):
    _grad_nnl = 0
    for i in range(len(y)):
        _grad_nnl += -y[i]/(1 + np.exp(y[i]*(X[:,[i]].T@_w)))*X[:,[i]]
    return _grad_nnl + 2*_w
```

```python
def gradient(w,l):
    return np.sum([-1 * y_train.T @ x_train / (1 + np.exp(-1 * y_train.T @ x_train @ w))]) + 2 * l * w

def gradient_descent(starting_w,l):
    w_current = starting_w
    tau = (2/(np.linalg.norm(x_train, 2)**2))/2
    for i in range(100):
        w_current += -1.0 * tau * gradient(w_current,l)
        
    return w_current
```

```python
XT = np.hstack((x_train, np.ones((n_train,1)))) #append a column of ones
X = XT.T
y = np.squeeze(y_train)

max_its = 5000
tau = .001

w = np.zeros((3,1)) #Convergence is sensitive to step size

## run gradient descent
w_new = w - tau*grad(w)
it = 1

while np.linalg.norm(w-w_new) > 0.001 and it < max_its:
    w = w_new
    w_new = w - tau*grad(w)
    it += 1
    #prints every 50 it
    if it%50 == 0:
        print("iteration:", it,", objective value:", regularized_log_loss(w))
        
print("w_opt:\n",w)
```

```python
plt.scatter(x_train[:,0],x_train[:,1] ,c=y_train[:,0])

#plot decision boundary

slope = -w[0,0]/w[1,0]
y_int = -w[2,0]/w[1,0]

plt.plot([0,1],[y_int,y_int + slope])
plt.title("Training data")
plt.show()

print("error rate:", np.mean(y_train != np.sign(X.T@w)))
```
